## 2.3 Load the stored data object

### About Tabula Muris Senis dataset
The published data objects have already been pre-processed and are essentially ready for the differential gene expression analysis and *etc.*
<br> Data objects available at: https://figshare.com/articles/dataset/Tabula_Muris_Senis_Data_Objects/12654728

<br><br>
**From A single-cell transcriptomic atlas characterizes ageing tissues in the mouse (2020)**
##### Data extraction 
 - Sequences from the NovaSeq de-multiplexed using bcl2fastq v.2.19.0.316
 - Reads  aligned to the mm10plus genome using STAR v.2.5.2b
 - Gene counts  produced using HTSEQ v.0.6.1p1 with default parameters, except stranded was set to false

##### Data pre-processing
 - Gene count tables were combined with the metadata variables using the Scanpy
 - Removed genes that were not expressed in at least 3 cells 
 - Removed cells that did not have at least 250 detected genes.
 - Removed cells with fewer than 5,000 counts
 - **Normalized** using size factor normalization such that every cell has 10,000 counts 
 - Log transformed
 - Computed highly variable genes (HVGs) and scaled the data to a maximum value of 10
 - Computed principal component analysis, neighbourhood graph and clustered the data using Louvain and Leiden methods, visualized using UMAP

##### Differential gene expression analysis
 - Well-powered sample size (more than 100 cells)
 - Used a linear model treating age as a numerical variable
 - Applied a false-discovery rate (FDR) threshold of 0.01 and an age coefficient threshold of 0.005 (corresponding to an approximately 10% fold change).


#### adata.Raw
 - Set the `.raw` attribute of the AnnData object to the normalized and logarithmized raw gene expression for later use in differential testing and visualizations of gene expression. This simply freezes the state of the AnnData object.
 - If you dont proceed below with correcting the data with `sc.pp.regress_out` and scaling it via `sc.pp.scale`, you can also **get away without using `.raw` at all**.

 - The result of the previous highly-variable-genes detection is stored as an annotation in `.var.highly_variable` and auto-detected by PCA and hence, sc.pp.neighbors and subsequent manifold/graph tools. In that case, the step actually do the filtering below is unnecessary, too.

````python
# Load the stored data object
adata = sc.read_h5ad('tabula-muris-senis-facs-processed-official-annotations-Brain_Non-Myeloid.h5ad')
adata.raw = adata
````

## 2.5 Subset data
Brain-Non-Myeloid data object contains more information than needed in this case (various cell types, female and male cells, and 3 different age groups...)

Choose just neurons from vars groups: neuron, interneuron, medium spiny neuron
Then subset the age groups 3m and 24m, lastly only select male cells (in many cases sex is a confounding variable in data) <br>


|**cells** |**3m data:**|**18m data:**|**24m data:**|
|---|------------|-------------|-------------|
|interneurons| 83 cells| 98 cells| 59 cells|
|medium spiny neurons| 89 cells|  11 cells| 3 cells|
|neurons| 154 cells| 123 cells| 207 cells|
|male| 266| 64| 269|
|female| 60| 168| 0|



`neuronlist = ["neuron", "interneuron", "medium spiny neuron"]` - creates a list datatype with names as they are in the `adata.obs["cell_ontology_class"]`
<br><br>
`adataneurons = adata[adata.obs["cell_ontology_class"].isin(neuronlist)]`
adataneurons = create a subset of datapoints that only have names from neuron list
<br><br>



````python
#subset the 3m and 24m age all neurons
#List of neuron types
neuronlist = ["neuron", "interneuron", "medium spiny neuron"]
# Splits the data into cell type groups
adataneurons = adata[adata.obs["cell_ontology_class"].isin(neuronlist)]

# List of neuron ages
neuron324 = ["3m", "24m"]
# Subsets ages 3m and 24m
adataneurons3_24 = adataneurons[adataneurons.obs["age"].isin(neuron324)]

# Male cells                               
neuronmale = ["male"]
# Subsets ages 3m and 24m
maleneurons324 = adataneurons3_24[adataneurons3_24.obs["sex"].isin(neuronmale)]
````

````python
#As seen from first line n_obs x n_vars now is smaller than it was in the adata,
#only selected 535 cells remain and all genes - 22966
maleneurons324
````

## 2.8 Marker Genes
Using the already clustered data we can identify genes that are differentially expressed in the clusters or groups. <br>
`rank_genes_groups()` function works well for a quick overview, however it should not be used for DGE analysis (use Diffxpy instead). Refer [here](https://github.com/theislab/scanpy/issues/397) for reasons behind. <br>

**Note:** Make use of `help()` function to get the details about certain functions <br>
There're also an option to look up individual functions on *scanpy.readthedocs* API section
[tl.rank_genes_groups](https://scanpy.readthedocs.io/en/stable/generated/scanpy.tl.rank_genes_groups.html#scanpy.tl.rank_genes_groups)
and
[pl.rank_genes_groups](https://scanpy.readthedocs.io/en/stable/generated/scanpy.pl.rank_genes_groups.html#scanpy.pl.rank_genes_groups).

`.tl.rank_genes_groups()` ranks the genes, whereas `.pl.rank_genes_groups()` - visualizes <br>
use `help(sc.tl.rank_genes_groups)` to find out about the parameters

* Note how first rank_genes_groups did not have `raference = 3m` and thus produced 2 graphs **3m vs rest** and **24m vs rest**, whereas the second had `reference = 3m` which set the "baseline" to 3m and only compared to it. It's not as important since there's only two age groups in the dataset, however you should be aware of this when analysing multiple groups (for example grouping by cell types), more examples in [here](https://nbisweden.github.io/workshop-scRNAseq/labs/compiled/scanpy/scanpy_05_dge.html#Compare-specific-clusters). <br>
* Also, note how two graphs have same y-axis, if you want to produce multiple ranking plots with their own y-axis use `sharey = False` in the `.pl`

````python
sc.tl.rank_genes_groups(maleneurons324, groupby='age', method = 't-test' )
sc.pl.rank_genes_groups(maleneurons324, groupby='age')
sc.pl.rank_genes_groups_dotplot(maleneurons324, key='rank_genes_groups', cmap='viridis')
sc.pl.rank_genes_groups_matrixplot(maleneurons324, key='rank_genes_groups', cmap='viridis')
sc.pl.rank_genes_groups_violin(maleneurons324, groups='24m', n_genes=3)
````

````python
sc.tl.rank_genes_groups(maleneurons324, groupby='age',reference='3m', method = 't-test' )
sc.pl.rank_genes_groups(maleneurons324, groupby='age')
sc.pl.rank_genes_groups_dotplot(maleneurons324, key='rank_genes_groups', cmap='viridis')
sc.pl.rank_genes_groups_matrixplot(maleneurons324, key='rank_genes_groups', cmap='viridis')
sc.pl.rank_genes_groups_violin(maleneurons324, groups='24m', n_genes=3)
````

Nonetheless, even though it's not recommended to use `rank_genes_groups` for DGE analysis, it's still possible to export the results in .csv

## 2.9 Diffxpy DGE analysis
Diffxpy documentation and tutorials can be found [here](https://diffxpy.readthedocs.io/en/latest/index.html)

````python
import diffxpy.api as de
import seaborn as sns
import logging
import scipy.stats
import matplotlib.pyplot as plt
````

#### Wald test

````python
#differential gene expression analysis between all male neuron types of 3m and 24m age groups
test_wald = de.test.wald(
    data=maleneurons324,
    formula_loc="~ 1 + age",
    factor_loc_totest="age"
)
````

We can then plot the DGE test results in a volcano plot. As you can see there is a sharp cut-off and then there are several dots that have -log(q-val)=30. This is because computer can't handle extremely precise numbers (anything smaller than 2.22e-16) and rounds numbers to 0 "rounded-up 0 p-val". Also, number of genes were only detected in only one age group (unsure whether it's biologically significant or just weren't detected during sequencing) "true 0 p-val". <br> 
You can distinguish if it's rounded up or true 0 by looking at log2FC. True 0 p-val had log2FC values -297 or 283.

````python
test_wald.plot_volcano(corrected_pval=True, min_fc=1.5, alpha=0.05, size=10) 
````

````python
testwaldres = test_wald.summary() #exporting results into a .csv file

testwaldres.to_csv(index=False)
compression_opts = dict(method='zip',
                       archive_name='testwaldresults.csv')  
testwaldres.to_csv('testwaldresults.zip', index=False,
         compression=compression_opts) 
````

#### T-test
The plot looks mirror image to Wald test volcano plot. The likely reason for this is how the functions read the age groups (alphabetically or numerically) and automatically assigns what's compared to what. When it's read alphabetically 24m comes before 3m (because 2 is lower than 3), when it's read numerically, the function recognises that 24m is larger than 3m. (at least that's the current likely reasoning, but might be something else entirelly)<br>
Wald test shows changes in 24m compared to 3m (3m set as baseline), whereas t-test compares how 3m differs from 24m (24m set as baseline).

````python
test_tt = de.test.t_test(
    data=maleneurons324,
    grouping="age"
)
````

````python
test_tt.plot_volcano(corrected_pval=True, min_fc=1.05, alpha=0.05, size=20)
````

We can still attempt to plot a scatterplot of p-values comparing between Wald and t-test.

````python
sns.scatterplot(
    x=test_wald.log10_qval_clean(),
    y=test_tt.log10_qval_clean()
)
plt.title('')
# Set x-axis label
plt.xlabel('Wald test -log10(q-values)')
# Set y-axis label
plt.ylabel('T-test -log10(q-values)')
````

#### Filtering csv file to contain only statistically significant results
Prior to running this unzip the results folder and move the .csv file out of results folder into main working directory folder "training"

````python
#Output csv file has only statistically significantly upregulated differentiated genes, 
#filtered out all the funky corner data, qval 0.05 and logfc smaller than 0.0585, 
#which is log fc 1.5
NeuronDGE = pd.read_csv('testwaldresults.csv')
upregfiltNeuronDGE = NeuronDGE.loc[
    (NeuronDGE['qval']<=0.05) &
    (NeuronDGE['log2fc']  < 283) & 
    (NeuronDGE['log2fc'] > 0.585)                         
                             ]
````

````python
#Output csv file has only statistically significantly downregulated differentiated genes, 
#filtered out all the funky corner data, qval 0.05 and logfc smaller than -0.0585, which is log fc 1.5
NeuronDGE = pd.read_csv('testwaldresults.csv')
downregfiltNeuronDGE = NeuronDGE.loc[
    (NeuronDGE['qval']<=0.05) &
    (NeuronDGE['log2fc'] > -297) & 
    (NeuronDGE['log2fc'] < -0.585)
]
    
````

````python
#merge upregulated and downregulated results that were filtered above into single .csv file
mergedfilteredDGE= pd.concat([downregfiltNeuronDGE, upregfiltNeuronDGE])
mergedfilteredDGE.to_csv(index=False)
compression_opts = dict(method='zip',
                        archive_name='Filtered DGE genes logfc and qval.csv')  
mergedfilteredDGE.to_csv('Filtered DGE genes logfc and qval.zip', index=False,
          compression=compression_opts) 
````
